### Description  
This notebook generates spatial visualizations to interpret the feasibility of submarine cable burial. It combines the operability constraints (depth and slope) with geomorphological clusters to highlight zones where burial is feasible or constrained. The outputs aim to support technical decision-making and portfolio-level communication.

### Inputs  
- `04_grid_operability.gpkg`: Grid with burial constraints and operability class  
- `03_grid_clusters_kmeans.gpkg`: Grid with `geomorpho_class`

### Processing  
- Merge operability and clustering data  
- Generate spatial maps for:  
  - Operability status  
  - Burial-required zones (highlighting infeasibility)  
  - Cross-tabulated view of clusters vs. burial feasibility  
- Generate histograms of depth and slope per cluster

### Outputs  
- `maps/05_operability_map.png`  
- `maps/05_burial_constraints_highlight.png`  
- `maps/05_cluster_vs_operability_table.png`

### Assumptions  
- Grid geometries are identical in both input files  
- All needed attributes are present and complete

```python
# Import libraries
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
# Load datasets
g1 = gpd.read_file("../processed_data/04_grid_operability.gpkg", layer="04_grid_operability")
g2 = gpd.read_file("../processed_data/03_grid_clusters_kmeans.gpkg", layer="03_grid_clusters_kmeans")
```

```python
# Merge datasets
grid = g1.merge(g2[["geometry", "geomorpho_class"]], on="geometry")
```

```python
# Define color maps
operability_colors = {
    "Required & Possible": "#78C679",
    "Required & Not Possible": "#D73027",
    "No Burial Needed": "#CCCCCC"
}
cluster_colors = ["#FFFFB2", "#FECC5C", "#FD8D3C", "#F03B20", "#BD0026"]
```

```python
# Plot operability map
fig, ax = plt.subplots(figsize=(10, 8))
grid.plot(column="operability_class", ax=ax, legend=True,
          categorical=True, cmap=plt.matplotlib.colors.ListedColormap(list(operability_colors.values())),
          edgecolor="black", linewidth=0.05)
plt.title("Burial Operability Map")
plt.axis("off")
plt.tight_layout()
plt.savefig("../outputs/maps/05_operability_map.png", dpi=300)
plt.show()
```

```python
# Highlight problematic areas
fig, ax = plt.subplots(figsize=(10, 8))
grid[grid["operability_class"] == "Required & Not Possible"].plot(ax=ax, color="#D73027", edgecolor="black", linewidth=0.05)
plt.title("Burial Required but Not Possible")
plt.axis("off")
plt.tight_layout()
plt.savefig("../outputs/maps/05_burial_constraints_highlight.png", dpi=300)
plt.show()
```

```python
# Histogram of slope and depth by cluster
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(x="geomorpho_class", y="slope_mean_deg", hue="geomorpho_class",
            data=grid, ax=axs[0], palette=cluster_colors, legend=False)
axs[0].set_title("Slope by Geomorphological Class")
axs[0].set_xlabel("Cluster")
axs[0].set_ylabel("Slope (deg)")

sns.boxplot(x="geomorpho_class", y="depth_mean", hue="geomorpho_class",
            data=grid, ax=axs[1], palette=cluster_colors, legend=False)
axs[1].set_title("Depth by Geomorphological Class")
axs[1].set_xlabel("Cluster")
axs[1].set_ylabel("Depth (m)")

plt.tight_layout()
plt.savefig("../outputs/maps/05_cluster_depth_slope_distributions.png", dpi=300)
plt.show()
```

```python
# Cross-tab summary
ctab = pd.crosstab(grid["geomorpho_class"], grid["operability_class"])
print("Cluster vs. Operability:\n")
print(ctab)
```
