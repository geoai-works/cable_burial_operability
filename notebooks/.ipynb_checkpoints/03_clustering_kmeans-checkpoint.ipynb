### Description
This notebook performs KMeans clustering on bathymetric and slope attributes to define geomorphological classes. The clusters allow segmentation of the study area based on seabed characteristics relevant for operational planning and cable route design.

### Inputs
- `02_grid_depth_slope.gpkg`: Grid with mean depth and slope per cell

### Processing
- Filter out grid cells with missing values
- Standardize variables
- Apply Elbow Method and Silhouette Score to select optimal K
- Apply KMeans clustering
- Assign class to each cell

### Outputs
- `03_grid_clusters_kmeans.gpkg`: Grid with geomorpho_class assigned per cell

### Assumptions
- Only cells with valid depth and slope are used
- Slope is used in degrees
- Clustering space is 2D: [depth_mean, slope_mean_pct]

### Dependencies
- geopandas, pandas, matplotlib, seaborn, sklearn, numpy

```python
# Import libraries
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.utils import resample
from sklearn.metrics import silhouette_score
```

```python
# Load grid with depth and slope
grid = gpd.read_file("../processed_data/02_grid_depth_slope.gpkg")
```

```python
# Drop rows with missing values
grid = grid.dropna(subset=["depth_mean", "slope_mean_deg"]).copy()
```

```python
# Select variables and scale
X = grid[["depth_mean", "slope_mean_deg"]].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Elbow Method
inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.axvline(x=6, color='red', linestyle='--', label='Selected k = 6')
plt.title("Elbow Method for K-Means")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("../outputs/plots/03_elbow_plot.png", dpi=300)
plt.show()
```

```python
# Silhouette Scores

# Sample 50,000 points from X_scaled for silhouette estimation
X_sample = resample(X_scaled, n_samples=50000, random_state=42)

# Recompute silhouette scores on the sample
silhouette_scores = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(X_sample)
    score = silhouette_score(X_sample, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o', color='green')
plt.axvline(x=6, color='red', linestyle='--', label='Selected k = 6')
plt.title("Silhouette Scores for K-Means")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("../outputs/plots/03_silhouette_scores.png", dpi=300)
plt.show()
```

### Justification for Selecting k = 6

Although the Elbow Method and Silhouette Score suggest that `k = 4` may be a statistically optimal choice, the selection of `k = 6` is retained based on technical and spatial considerations specific to submarine cable planning.

The main reasons for preferring `k = 6` are:

- **Improved morphological differentiation**: With `k = 6`, the clusters better separate key geomorphological features such as escarpments, slope breaks, and fine-scale topographic transitions that are critical in cable route assessments.
  
- **Enhanced correspondence with slope patterns**: When overlaid on the slope raster, the `k = 6` clustering captures high-slope zones (operationally relevant) as distinct classes, which are not represented adequately with `k = 4`.

- **Spatial coherence and interpretability**: The resulting clusters in `k = 6` exhibit consistent and meaningful spatial distribution that aligns with known seabed structures and operational constraints.

- **Decision-oriented utility**: The added complexity from `k = 6` provides more actionable segmentation without overfitting, allowing for better risk-based route planning and prioritization.

> In summary, `k = 6` is chosen not solely on statistical grounds, but on the added value it brings in operational insight and geospatial structure recognition.

```python
# Apply final KMeans with k=6
kmeans = KMeans(n_clusters=6, random_state=42)
grid["geomorpho_class"] = kmeans.fit_predict(X_scaled)
```

```python
# Save output
output_path = "../processed_data/03_grid_clusters_kmeans.gpkg"
grid.to_file(output_path, layer="clusters", driver="GPKG")

print(f"Clustering complete. Saved to: {output_path}")
```

```python
# Plot clusters on map

# Define a color palette with 6 distinguishable colors
cluster_colors = {
    0: "#1f78b4",
    1: "#33a02c",
    2: "#e31a1c",
    3: "#ff7f00",
    4: "#6a3d9a",
    5: "#b15928"
}

# Create color column
grid["color"] = grid["geomorpho_class"].map(cluster_colors)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))
grid.plot(color=grid["color"], linewidth=0, ax=ax)

# Create custom legend
import matplotlib.patches as mpatches
handles = [
    mpatches.Patch(color=color, label=f"Class {i}")
    for i, color in cluster_colors.items()
]
ax.legend(handles=handles, title="Geomorpho Class", loc="upper right")

plt.title("Geomorphological Classes (K-Means Clustering, k=6)")
plt.axis("off")
plt.tight_layout()
plt.savefig("../outputs/plots/03_cluster_map.png", dpi=300)
plt.show()
```
