### Description  
This notebook evaluates the operational risk associated with simulated submarine cable routes. Using the previously classified cell-level operational complexity, it intersects each route with the grid and assigns a risk score and category. It outputs a summary table and visual diagnostics.

### Inputs  
- `06_grid_cell_complexity.gpkg`: Grid with slope, depth and cell-level complexity classification  
- `aoi.gpkg`: Area of interest polygon  
- `simulated_routes.gpkg`: Three manually-designed routes in QGIS (fields: `fid = 1, 2, 3`)

### Processing  
- Intersect routes with operational complexity grid  
- Assign risk score per cell (Low = 0, Moderate = 1, High = 2, Critical = 3)  
- Aggregate risk by route and classify total score  
- Output GeoPackage with segmented risk and summary CSV  
- Generate bar chart and route complexity map

### Outputs  
- `07_route_operational_risk.csv`  
- `07_routes_with_risk.gpkg`  
- `07_route_risk_scores.png`  
- `07_route_complexity_map.png`

```python
# Import libraries
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib.colors import ListedColormap
```

```python
# Load input data
grid = gpd.read_file("../processed_data/06_grid_cell_complexity.gpkg", layer="06_grid_cell_complexity")
routes = gpd.read_file("../qgis_project/simulated_routes.gpkg")
```

```python
# Ensure same CRS
if grid.crs != routes.crs:
    routes = routes.to_crs(grid.crs)
```

```python
# Add an explicit route_id field if it doesn't exist
routes = routes.copy()
routes["route_id"] = routes.index + 1
```

```python
# Intersect routes with grid and retain all geometries
intersections = gpd.overlay(grid, routes, how="intersection", keep_geom_type=False)
```

```python
# Assign route ID and risk points
intersections["route_id"] = intersections["route_id"]
risk_points = {
    "Low": 0,
    "Moderate": 1,
    "High": 2,
    "Critical": 3
}
intersections["risk_score"] = intersections["cell_complexity"].map(risk_points)
intersections["length_m"] = intersections.geometry.length
```

```python
# Summarize risk per route
route_summary = intersections.groupby("route_id").agg(
    total_length=("length_m", "sum"),
    mean_risk_score=("risk_score", "mean"),
    total_risk_score=("risk_score", "sum"),
    low_pct=("cell_complexity", lambda x: 100 * (x == "Low").sum() / len(x)),
    mod_pct=("cell_complexity", lambda x: 100 * (x == "Moderate").sum() / len(x)),
    high_pct=("cell_complexity", lambda x: 100 * (x == "High").sum() / len(x)),
    crit_pct=("cell_complexity", lambda x: 100 * (x == "Critical").sum() / len(x)),
    n_cells=("geometry", "count")
).reset_index()
```

```python
# Classify overall route risk
def classify_route(score):
    if score <= 10:
        return "Low"
    elif score <= 30:
        return "Moderate"
    elif score <= 60:
        return "High"
    else:
        return "Critical"

route_summary["route_risk_level"] = route_summary["total_risk_score"].apply(classify_route)
```

```python
# Save outputs
os.makedirs("../outputs/summary", exist_ok=True)
os.makedirs("../outputs/maps", exist_ok=True)
os.makedirs("../processed_data", exist_ok=True)

route_summary.to_csv("../outputs/summary/07_route_operational_risk.csv", index=False)
intersections.to_file("../processed_data/07_routes_with_risk.gpkg", layer="07_routes_with_risk", driver="GPKG")
print("Outputs saved: CSV summary and GPKG routes")
```

```python
# Bar chart of total risk scores
plt.figure(figsize=(8, 5))
sns.barplot(data=route_summary, x="route_id", y="total_risk_score", hue="route_risk_level", dodge=False,
            palette={"Low": "#91cf60", "Moderate": "#fee08b", "High": "#fc8d59", "Critical": "#d73027"})
plt.title("Operational Risk Score by Route")
plt.xlabel("Route ID")
plt.ylabel("Total Risk Score")
plt.tight_layout()
plt.savefig("../outputs/maps/07_route_risk_scores.png", dpi=300)
plt.show()
```

```python
# Map segments by complexity with ordered legend
ordered_classes = ["Critical", "High", "Moderate", "Low"]
colors = {
    "Critical": "#d73027",
    "High": "#fc8d59",
    "Moderate": "#fee08b",
    "Low": "#91cf60"
}
cmap = ListedColormap([colors[c] for c in ordered_classes])
intersections["cell_complexity"] = pd.Categorical(intersections["cell_complexity"],
                                                  categories=ordered_classes,
                                                  ordered=True)

fig, ax = plt.subplots(figsize=(10, 8))
intersections.plot(column="cell_complexity", ax=ax, cmap=cmap, legend=True, linewidth=2)
plt.title("Route Segments by Operational Complexity")
plt.axis("off")
plt.tight_layout()
plt.savefig("../outputs/maps/07_route_complexity_map.png", dpi=300)
plt.show()
```

```python
# Export route segments by complexity for QGIS visualization
segments_for_qgis = intersections[["route_id", "cell_complexity", "geometry"]].copy()

# Optional: enforce ordered category for legend consistency
segments_for_qgis["cell_complexity"] = pd.Categorical(
    segments_for_qgis["cell_complexity"],
    categories=["Critical", "High", "Moderate", "Low"],
    ordered=True
)

segments_for_qgis.to_file(
    "../processed_data/07_route_segments_by_complexity.gpkg",
    layer="segments_by_complexity",
    driver="GPKG"
)

print("Exported: 07_route_segments_by_complexity.gpkg (ready for QGIS)")
```
